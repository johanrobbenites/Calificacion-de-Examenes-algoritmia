```python
texto = (input("Ingrese el texto: ").lower()).split(" ")
def palindromo(texto, inicio, fin):
    if inicio >= fin:
        return True
    if texto[inicio] == texto[fin]:
        return palindromo(texto, inicio + 1, fin - 1)
    else:
        return False
print(palindromo(texto, 0, -1))
#True: es palindromo
#False: no es palindromo
```
