```python
archivo = open("Alumnos.txt","r")
alumnos=[]
cursos=[]
cont=0
cont_1=0
for i in archivo:
  y = i.split("\n")
  x = y[0].split(" ")
  cod_al=x[0]
  cod_cur=x[1]
  nota=float(x[2])
  e=True
  if cont ==0:
    if nota>=10:
      r=[cod_al,1,0]
    else:
      r=[cod_al,0,1]
    alumnos.append(r)
    cont=+1
    cont_1=+1
    r1=[cod_cur,0]
    cursos.append(r1)
  else:
    for j in range(cont):
      if cod_al==alumnos[j][0]:
        e=False
        if nota>=10:
          alumnos[j][1]=alumnos[j][1]+1
        else:
          alumnos[j][2]=alumnos[j][2]+1
      else:
        if (j+1==cont) and (e==True):
          r=[cod_al,0,0]
          alumnos.append(r)
          if nota>=10:
            alumnos[j+1][1]=alumnos[j+1][1]+1
          else:
            alumnos[j+1][2]=alumnos[j+1][2]+1
          cont=cont+1
        

  if nota<=10 and cont !=0:
    for z in range(cont_1):
      if cod_cur==cursos[z][0]:
        cursos[z][1]=+1
      else:
        r1= [cod_cur,1]
        cursos.append(r1)
    cont_1 = +1
for i in range(len(alumnos)):
  print(f"Alumno de codigo {alumnos[i][0]} tiene {alumnos[i][1]} cursos aprobados y {alumnos[i][2]} cursos desaprobados")
for i in range(len(cursos)):
  print(f"El curso {cursos[i][0]} tiene {cursos[i][1]} desaprobados")
archivo.close()
```
