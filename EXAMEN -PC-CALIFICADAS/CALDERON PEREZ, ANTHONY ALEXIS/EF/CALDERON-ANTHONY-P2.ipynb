```python
f=open('/content/drive/MyDrive/CALDERON PEREZ, ANTHONY ALEXIS/EF/Archivo.txt',"r")

v=[]
n=[]
for i in f:
  reg=i.split(" ")
  venta=float(reg[0])
  nombre=reg[1]+" " +reg[2]
  v.append(venta)
  n.append(nombre)


for i in range(len(v)):
  for j in range(len(v)):
    if (j!=(len(v)-1)):
      if (v[j]<v[j+1]):
        a=v[j]
        v[j]=v[j+1]
        v[j+1]=a
        e=n[j]
        n[j]=n[j+1]
        n[j+1]=e
pos=[]
for i in range(len(v)):
  if i==0:
    pos.append(i+1)
  else:
    pos.append(i+1)
    if v[i]==v[i-1]:
      pos[i]=pos[i-1]

print("Monto      Nombre            Pto        Bono           Porc")
for i in range(len(v)):
  por=11-pos[i]
  vent=v[i]
  b=(por/100)*vent
  print(f"{v[i]}    {n[i]}", end=(" ")*(19-len(n[i])))
  print(f"{pos[i]}         ",end="")
  print("{:.2f}".format(b),f"       {por}%")
 

f.close()
```
