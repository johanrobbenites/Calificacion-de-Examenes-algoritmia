```python
def invertir(ora,orain,x):
  if x==len(ora):
    print(orain)
    if ora==orain:
      print("La oracion es palindroma")
    else:
      print("La oracion no es palindroma")
  else:
    orain=orain+ora[((len(ora))-1)-x]
    invertir(ora,orain,x+1)

oracion=input("Ingrese oración: ")
for i in range(len(oracion)):
  oracion=oracion.lower()

l=[]
for i in range(len(oracion)):
  if oracion[i]!=" ":
    l.append(oracion[i])
ora=""
for i in range(len(l)):
  ora=ora+l[i]
  
orain=""
x=0



print("Oracion analizada inicial:",ora)
print("Oracion analizada invertida: ", end="" )
invertir(ora,orain,x)


```
