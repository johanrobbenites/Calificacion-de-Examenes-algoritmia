```python
l=[]
a=int(input("Ingrese el numero de examenes: "))
for i in range(a):
  n=float(input(f"Ingrese la nota del examen {i+1}: "))
  l.append(n)
p=0
while(p<0.7):
  k=0
  for i in l:
    print(i, end=" ")
    if (i >=10):
      
      k=k+1
      print(k, end=" ")
  p=(k/a)
  print(f" {p*100} % de aprobados\n")
  for i in range(a):
    if (19>=i):
      l[i]=l[i]+1
```
