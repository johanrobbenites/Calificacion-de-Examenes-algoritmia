```python
def datos(n,nom,c,p):
  for i in range(n):
    while(True):
      codigo=input(f"codigo postulante {i+1}: ")
      if (len(codigo)==5):
        break
      else:
        print("Codigo invalido")
    nombre=input(f"nombre postulante {i+1}: ")
    while(True):
      puntaje=float(input(f"puntaje postulante {i+1}: "))
      if puntaje>=0 and puntaje<=20:
        break
      else:
        print("Codigo invalido")
    nom.append(nombre)
    c.append(codigo)
    p.append(puntaje)

def ordenar(n,nom,c,p):
  nom.append(0)
  c.append(0)
  p.append(0)
  for k in range(n):
    max=0
    for y in range(k,n):
      if max <= p[y]:
        if max < p[y]:
          max= p[y]
          ind=y
    p[n]=p[k]
    p[k]=p[ind]
    p[ind]=p[n]
    c[n]=c[k]
    c[k]=c[ind]
    c[ind]=c[n]
    nom[n]=nom[k]
    nom[k]=nom[ind]
    nom[ind]=nom[n]
  nom.pop()
  c.pop()
  p.pop()

def imprimir(n,nom,c,p):
  print("CODIGO  NOMBRE               PUNTAJE   PUESTO")
  for i in range(n):
    print(f"{c[i]}  {nom[i]}      {p[i]}   {i}")

def main():
  n=int(input("Ingrese numero de postulantes: "))
  nom=[]
  c=[]
  p=[]
  datos(n,nom,c,p)
  ordenar(n,nom,c,p)
  imprimir(n,nom,c,p)
main()
```
