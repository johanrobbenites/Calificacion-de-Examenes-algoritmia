```python
def datos(n,cads):
  print("Ingrese cadenas (<50 caracteres)")
  for i in range(n):
    while(True):
      c=input(f"Ingrese cadena {i+1}: ")
      if len(c)<=50:
        break
      else:
        print("Numero de caracteres invalido")
    cads.append(c)

def caracteres(n,cads):
  alf=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','s','t','u','v','w','x','y','z']
  for i in range(len(alf)):
    x=0
    for j in range(len(cads)):
      for k in cads[j]:
        if alf[i]==k:
          x=x+1
    if x>0:
      print(f"{alf[i]}: {x}", "*"*x)

def main():
  cads=[]
  n=int(input("Ingrese el numero de cadenas: "))
  datos(n,cads)
  caracteres(n,cads)

main()
```
