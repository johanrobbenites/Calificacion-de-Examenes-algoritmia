```python
def elementos(n,c,l):
  print("Conjunto", c)
  for i in range(n):
    x=int(input(f"Elemento {i+1}:"))
    l.append(x)
  print("")

def ordenar(l):
  n=len(l)
  l.append(0)
  max=0
  for i in range(n):
    if l[i]>max:
      max=l[i]
  for k in range(n):
    min=max+1
    for y in range(k,n):
      if min >= l[y]:
        if min > l[y]:
          min= l[y]
          ind=y
    l[n]=l[k]
    l[k]=l[ind]
    l[ind]=l[n]
  l.pop()

def imprimir(l,c):
  print(f"{c} =", "{" ,end="")
  for i in range(len(l)):
    if i!=len(l)-1:
      print(f"{l[i]}; ",end="")
    else:
      print(l[i],end= "}")
  print("")

def inter(it,m,n):
  for i in range(len(m)):
    for j in range(len(n)):
      if m[i]==n[j]:
        it.append(m[i])

def union(u,m,n):
  for i in m:
    u.append(i)
  for i in n:
    u.append(i)
def main():
  m=[]
  n=[]
  u=[]
  it=[]
  a="A"
  b="B"
  uni="UNION"
  intr="INTERSECCION"
  n1=int(input("Cantidad de elementos del conjunto A: "))
  n2=int(input("Cantidad de elementos del conjunto B: "))
  elementos(n1,a,m)
  elementos(n2,b,n)
  ordenar(m)
  ordenar(n)
  imprimir(m,a)
  imprimir(n,b)
  inter(it,m,n)
  union(u,m,n)
  ordenar(u)
  imprimir(u,uni)
  imprimir(it,intr)

main()
```
