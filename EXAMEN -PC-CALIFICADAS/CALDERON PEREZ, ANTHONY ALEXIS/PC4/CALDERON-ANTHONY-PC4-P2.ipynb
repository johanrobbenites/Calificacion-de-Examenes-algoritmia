```python
f=open("ALUMNOS.txt","r")
c=[]
n=[]
no=[]
k=0
for i in f:
  reg=i.split(" ")
  cod=reg[0]
  nom=reg[1]
  nota=reg[2]
  c.append(cod)
  n.append(nom)
  no.append(nota)

  k=k+1


print("CODIGO  CURSO     PUNTAJE  ESTADO")
for i in range(k):
  print(f"{c[i]}     {n[i]}     {no[i]}",end="     ")
  if float(no[i])>=10:
    print("Aprobado")
  else:
    print("Desaprobado")


f.close()
```
