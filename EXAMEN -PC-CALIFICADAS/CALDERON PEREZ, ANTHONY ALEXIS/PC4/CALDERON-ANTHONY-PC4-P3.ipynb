```python
def buscar(l,dato,i,x):
    if l[i]==dato:
        print(f"Posición del dato: {i}")
    elif (len(l)-1)==i and (x==-1):
        print(f"Posición del dato: {x}")
    
    else:
        buscar(l,dato,i+1,x)
    

def main():
  n=int(input("Ingrese cantidad de numeros: "))
  l=[]
  for i in range(n):
    x=int(input(f"Numero {i+1}: "))
    l.append(x)
  print(l)
  dato=int(input("Ingrese dato a buscar: "))
  buscar(l,dato,0,-1)

main()
```
