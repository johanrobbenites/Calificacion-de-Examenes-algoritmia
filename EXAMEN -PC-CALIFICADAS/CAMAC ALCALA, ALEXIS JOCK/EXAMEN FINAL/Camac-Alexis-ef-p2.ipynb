```python
def chartonum(l):
    flag,num,numdecimal,c=0,0,0,0
    for i in l:
        if i==".":
            flag=1
        if (ord(i)>=48 and ord(i)<=57) and flag==0:
            num=num*10+ord(i)-48
        if (ord(i)>=48 and ord(i)<=57) and flag==1:
            numdecimal=numdecimal*10+ord(i)-48
            c+=1
    if c==2:
        num=num+numdecimal*0.01
    if c==1:
        num=num+numdecimal*0.1
    return num

f=open("registronombres.txt","r+")
l=f.read()
l=l.splitlines()
montos=[]
nombres=[]
j=0
for i in l:
    linea=i.split()
    montos.append(chartonum(linea[0]))
    nombres.append(linea[1]+" "+linea[2])

#Ordenando montos:
for i in range(1,len(montos)):
    for j in range(len(montos)-i):
        if(montos[j]<montos[j+1]):
            (montos[j],montos[j+1])=(montos[j+1],montos[j])
            (nombres[j],nombres[j+1])=(nombres[j+1],nombres[j])

pos=[1]
aux=1
for i in range(1,len(montos)):
    if montos[i-1]>montos[i]:
        aux+=1
    pos.append(aux)

c=0
for i in range(len(pos)):
    if pos[i]<=5:
        c+=1
bonos=[]
porcentajes=[]
for i in range(len(montos)):
    if c!=0:
        por=11-pos[i]
        porcentajes.append(por)
        bono=(0.01)*(11-pos[i])*montos[i]
        bonos.append(round(bono,2))
        c-=1
    if c==0:
        porcentajes.append('-')
        bonos.append('-')

for i in range(len(montos)):
    print(str(montos[i])+" "+nombres[i]+" "*(20-len(nombres[i]))+str(pos[i])+" "+str(bonos[i])+" "*(10-len(str(bonos[i])))+str(porcentajes[i])+"%")
```
