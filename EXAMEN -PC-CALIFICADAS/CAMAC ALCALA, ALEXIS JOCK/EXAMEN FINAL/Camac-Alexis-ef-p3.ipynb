```python
def sin_espacio(oracion):
    new=""
    for i in oracion:
        if i!=" ":
            new=new+i
    return new
def revertir(oracion):
    oracion=sin_espacio(oracion)
    if oracion == "":
        return oracion
    else:
        return oracion[0]+revertir(oracion[1:])

oracion=input("Ingrese la oracion: ")
oracion=oracion.lower()
oracion=sin_espacio(oracion)
if oracion==revertir(oracion):
    print("La oracion es palindroma")
    print(oracion+" = "+revertir(oracion))
else:
    print("La oracion no es palindroma")
```
