```python

def convertiraentero(l):
    dec=0
    num=0
    numdecimal=0
    index=0
    for i in l:
        if i==",":
            dec=1
        if (ord(i)>=48 and ord(i)<=57) and dec==0:
            num=num*10+ord(i)-48
        if (ord(i)>=48 and ord(i)<=57) and dec==1:
            numdecimal=numdecimal*10+ord(i)-48
            index+=1
    if index==1:
        num=num+numdecimal*0.1
    return num

f=open("alumnos.txt","r")
l=f.read()
l=l.splitlines()
codigos=[]
cursos=[]
for i in l:
    linea=i.split()
    if linea[0] not in codigos:
        codigos.append(linea[0])
reg=[]
#REGISTRO DE ALUMNOS
for i in range(len(codigos)):
    ap=0
    des=0
    for j in l:
        linea=j.split()
        if linea[0]==codigos[i]:
            if convertiraentero(linea[2])>=10:
                ap+=1
            else:
                des+=1
    reg.append([codigos[i],ap,des])
for i in range(len(reg)):
    print(f'Codigo de alumno {reg[i][0]} : ')
    print(f'Cursos aprobados: {reg[i][1]} ')
    print(f'Cursos desaprobados: {reg[i][2]} ')

#BUSCANDO EL CURSO...
for i in l:
    linea=i.split()
    if linea[1] not in cursos:
        cursos.append(linea[1])
max=0
for i in range(len(cursos)):
    ap=0
    des=0
    for j in l:
        linea=j.split()
        if linea[1]==cursos[i]:
            if convertiraentero(linea[2])>=10:
                ap+=1
            else:
                des+=1
    if(max<des):
        max=des
        cursobuscado=cursos[i]
print("Curso cuyo numero de desaprobados es el mayor a todos:  ",cursobuscado)
```
