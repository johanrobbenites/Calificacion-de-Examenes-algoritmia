```python
def buscando(lista, i, j, valor):
    if j >= i:
        medio = (j + i) // 2
        if lista[medio] == valor:
            return medio
        elif lista[medio] > valor:
            return buscando(lista, i, medio-1, valor)
        else:
            return buscando(lista, medio+1, j, valor)
    else:
        return -1

n=int(input("Ingrese la cantidad de numeros: "))
x=[]
for i in range(n):
    num=int(input(f'Ingrese el valor #{i+1} : '))
    x.append(num)
valor=int(input("Ingrese el numero a buscar: "))

result = buscando(x, 0, len(x)-1, valor)
 
if result != -1:
    print("El elemento esta en la posicion: ", str(result))
else:
    print("El elemento no esta en la lista")
```
