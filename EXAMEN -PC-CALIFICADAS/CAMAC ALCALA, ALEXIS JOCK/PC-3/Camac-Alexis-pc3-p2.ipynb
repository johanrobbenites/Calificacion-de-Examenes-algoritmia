```python
index=[]
#lista creada
for i in range(97,123):
    index.append(0)
n=int(input("Ingrese cuantas cadenas usara: "))
for i in range(n):
    o=input(f'Ingrese la cadena #{i+1}: ')
    while(len(o)>50):
      o=input(f'Ingrese la cadena nuevamente: ')
    for j in o:
        if j!='á' or j!="é" or j!="í" or j!="ó" or j!="ú":
            if ord(j)>=97 and ord(j)<=122:
                index[ord(j)-97]+=1
for i in range(len(index)):
    if(index[i]!=0):
        print(chr(i+97),index[i],'*'*index[i])
```
