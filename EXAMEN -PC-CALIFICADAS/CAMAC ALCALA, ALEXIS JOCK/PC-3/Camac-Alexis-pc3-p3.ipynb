```python
a="A={-41;-2;0;1;123}"
b="B={-3;-2;1;2}"
#MAYUSCULAS= 65 AL 90 (A,B,C....Z)
#NUMEROS=  48 AL 57 (0,1,2.....9)
#MINISCULAS= 97 AL 122(a,b,c,d.... z)

#transformando a numeors
def conjuntos(a,b):
  lista1=[]
  lista2=[]
  # '-'=45
  signo=1
  num=0
  for i in a:
      if(i!=';'):
          if(ord(i)==45):
              signo=-1
          if(ord(i)>=48 and ord(i)<=57):
              num=num*10+ord(i)-48
      if(i==';' or i=='}'):
          num=num*signo
          signo=1
          lista1.append(num)
          num=0
  for i in b:
      if(i!=';'):
          if(ord(i)==45):
              signo=-1
          if(ord(i)>=48 and ord(i)<=57):
              num=num*10+ord(i)-48
      if(i==';' or i=='}'):
          num=num*signo
          signo=1
          lista2.append(num)
          num=0

  inter=[]
  #interseccion:
  for i in range(len(lista1)):
      for j in range(len(lista2)):
          if(lista1[i]==lista2[j]):
              if lista1[i] not in inter:
                  inter.append(lista1[i])

  #union:
  uni=lista1+lista2
  union=[]
  for i in range(len(uni)):
      if uni[i] not in union:
          union.append(uni[i])
  for i in range(1,len(union)):
      for j in range(len(union)-i):
          if(union[j]>union[j+1]):
              (union[j],union[j+1])=(union[j+1],union[j])
  cad1=""
  cad2=""
  #MAYUSCULAS= 65 AL 90 (A,B,C....Z)
  #NUMEROS=  48 AL 57 (0,1,2.....9)
  #MINISCULAS= 97 AL 122(a,b,c,d.... z)
  for i in union:
      cad1 = cad1 + " " + str(i) + ";"
  cad1 = "{"+cad1[1:-1]+"}"
  print("UNION = " + cad1)

  for i in inter:
      cad2 = cad2 + " "+ (str(i) + ";")
  cad2 = "{"+cad2[1:-1]+"}"
  print("INTERSECCION = " + cad2)
conjuntos(a,b)

```
