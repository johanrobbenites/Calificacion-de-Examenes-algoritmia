```python
n = int(input('Ingrese valor de "n": '))
diccionary = {}
def ingresar_codigo():
    x = str(input('Ingrese código: '))
    if not x.isalnum():
        print('El código debe tener caracteres alfanuméticos.')
        ingresar_codigo()
    elif len(x) != 5 :
        print('El código debe tener 5 caracteres')
        ingresar_codigo()
    else:
        return x

def ingresar_nombre():
    x = str(input('Ingrese código'))
    if not x.isalpha():
        print('El nombre debe tener caracteres alfabetos.')
        ingresar_nombre()
    elif len(x) > 50 :
        print('El nombre debe tener hasta 50 caracteres')
        ingresar_nombre()
    else:
        return x

def ingresar_puntaje():
    x = float(input('Ingrese código'))
    if not x.isdigit():
        print('El puntaje debe tener caracteres dígitos.')
        ingresar_puntaje()
    elif x > 20 and len(str(x)) > 5 :
        print('El puntaje debe ser menor que 20 y con maximo 3 decimales')
        ingresar_puntaje()
    else:
        return x

for i in range(n):
    codigo = ingresar_codigo()
    nombre = ingresar_nombre()
    puntaje = ingresar_puntaje()
    diccionary[codigo]=[nombre,puntaje]

```
