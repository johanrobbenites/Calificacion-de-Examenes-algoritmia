```python
def union_interseccion(A,B):
    total=[]
    for conjunto in [A,B]:
        conjunto=conjunto.strip('{}')
        lista=conjunto.split(';')
        total.extend(lista)

    indices=[]
    for element in total:
        contador = total.count(element)
        if contador > 1:
            indices.append(total.index(element))
    
    interseccion = []
    for i in indices:
        interseccion.append(total.pop(i))
    
    union = total

    burbuja(union)
    burbuja(interseccion)
    return union , interseccion


def burbuja(lista):
    change = True
    while change:
        change = False
        for i in range(len(lista)-1):
            if lista[i] > lista[i+1]:
                lista[i], lista[i+1] = lista[i+1], lista[i]
                change = True
    return lista


A = '{-41;-2;0;1;123}'
B = '{-3;-2;1;2}'
UNION, INTERSECCION = union_interseccion(A,B)
print(UNION)
print(INTERSECCION)

```
