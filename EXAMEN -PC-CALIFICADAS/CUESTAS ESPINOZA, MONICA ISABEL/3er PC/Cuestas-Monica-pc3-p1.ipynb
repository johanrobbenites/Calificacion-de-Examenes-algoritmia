```python
nombres=[]
notas=[]
codigos=[]
n=int(input("escriba el numero de alumnos: "))
for i in range(n):
    codi=str(input("código del alumno "+str(i+1)+" : "))
    while(len(codi)!=5):
        codi=str(input("Ingrese un código de 5 caracteres: "))
    codigos.append(codi)

    name=str(input("Ingrese el nombre del postulante "+str(i+1)+" : "))
    while(len(name)>50):
        name=str(input("ingrese de nuevo: "))
    nombres.append(name)

    nota=float(input("Ingrese la nota del alumno "+str(i+1)+" : "))
    while(nota > 20 or nota < 0):
        nota = float(input("nota ente 0 y 20"))
    notas.append(nota)
x=[]
for i in range(n):
    x.append([0]*3)
for j in range(3):
    for i in range(n):
        if(j==0):
            x[i][j]=codigos[i]
        if(j==1):
            x[i][j]=nombres[i]
        if(j==2):
            x[i][j]=notas[i]
for i in range(n):
    for j in range(3):
        print(x[i][j],end=" ")
    print()
#usando el metodo burbuja
for i in range(1,len(notas)):
    for j in range(len(notas)-i):
        if(notas[j]<notas[j+1]):
            m=notas[j]
            notas[j]=notas[j+1]
            notas[j+1]=m
            n=codigos[j]
            codigos[j]=codigos[j+1]
            codigos[j+1]=n
            y=nombres[j]
            nombres[j]=nombres[j+1]
            nombres[j+1]=y


for j in range(3):
    for i in range(n):
        if(j==0):
            x[i][j]=codigos[i]
        if(j==1):
            x[i][j]=nombres[i]
        if(j==2):
            x[i][j]=notas[i]
```
