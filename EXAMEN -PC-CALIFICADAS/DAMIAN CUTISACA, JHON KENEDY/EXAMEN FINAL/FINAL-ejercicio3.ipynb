```python
cadena = str(input("Escribe una oración : "))
c = []
b = ""
d = ""
def elimin(text,c):
  m= text.lower()
  for i in m:
    a = str(i)
    if a != " ":
      c.append(a)
print("La oración ingresada es :")
print(cadena)
elimin(cadena,c)
for i in range(len(c)):
  b = b + c[i]
print("Sin espacios en blanco y todo en minúsculas sería:")
print(b)
copia = []
for i in c:
  a = i
  copia.append(a)
#Función para invertir la oración usando recursividad
def invertir(cad,i):
      if i == (len(cad)//2):
        return cad
      else :  
        a = cad[len(cad)-i-1]
        cad[len(cad)-i-1] = cad[i]
        cad[i] = a
        invertir(cad,i+1)
      cad[i] = a
invertir(c,0)
print("Al invertirla da :")
for i in range(len(c)):
  d = d + c[i]
print(d)
if c == copia:
  print("Por tanto,es palíndromo")
else :
  print("Por tanto,no es palindromo")
```
