```python
filas = int(input("NUMERO DE FILAS :"))
columnas = int(input("NUMERO DE COLUMNAS :"))
matriz = [ ]
for i in range(filas):
  matriz.append([])
  for j in range(columnas) :
      valor = int(input("Fila {} , Columna {} :".format(i+1,j+1)))
      matriz[i].append(valor)
for g in range(len(matriz)):
  print(matriz[g])

rotar1 = []
for n in range(len(matriz[0])):
    rotar1.append([])
    for m in range(len(matriz)):
        rotar1[n].append(matriz[len(matriz)-1 - m][n])
for g1 in range(len(rotar1)):
  print(rotar1[g1])

rotar2 = []
for n in range(len(rotar1[0])):
    rotar2.append([])
    for m in range(len(rotar1)):
        rotar2[n].append(rotar1[len(rotar1)-1 - m][n])
for g2 in range(len(rotar2)):
  print(rotar2[g2])

rotar3 = []  
for n in range(len(rotar2[0])):
    rotar3.append([])
    for m in range(len(rotar2)):
        rotar3[n].append(rotar2[len(rotar2)-1 - m][n])
for g3 in range(len(rotar3)):
  print(rotar3[g3])
rotar4 = []  
for n in range(len(rotar3[0])):
    rotar4.append([])
    for m in range(len(rotar3)):
        rotar4[n].append(rotar3[len(rotar3)-1 - m][n])
for g4 in range(len(rotar4)):
  print(rotar4[g4])
```
