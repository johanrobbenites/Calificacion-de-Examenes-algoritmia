```python
nalum = int(input("ingresar alumnos : "))
listnotas = []
aprob = 0
for i in range(0,nalum):
   listnotas.append(input(" "))
   if int(listnotas[i]) >= 10:
    aprob = aprob + 1
for j in range(0,nalum):
   print(listnotas[j], end = " ")
print((aprob/nalum)*100 , "% de aprobados") 
if(aprob/nalum) < 0.5:
  while (aprob/nalum) < 0.5:
    aprob = 0
    for j in range(0,nalum):
      (listnotas[j]) = int(listnotas[j]) + 1
    for h in range(0,nalum):
       print(listnotas[h], end = " ")
       if int(listnotas[h]) >= 10:
          aprob = aprob + 1
    print((aprob/nalum)*100 , "% de aprobados") 
    

```
