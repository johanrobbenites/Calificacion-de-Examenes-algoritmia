```python
n = int(input("CANTIDAD DE NUMEROS QUE INGRESARÁ :"))
lista = []
for i in range(n):
  num = int((input("INGRESE NUMERO : ")))
  lista.append(num) 
def buscar(a,b,c,m):
  if m == c:
    return -1
  if a == b[m]:
    return m
  else :
   return buscar(a,b,c,m+1)
elem = int(input("SOLICITE UN NUMERO :"))
print("LA POSICION QUE APARECE EL NUMERO SOLICITADO ES :" ,buscar(elem,lista,n,0) )
```
