```python
def chartonum(l):
    dec=0
    num=0
    numdec=0
    c=0
    for i in l:
        if i==",":
            dec=1
        if (ord(i)>=48 and ord(i)<=57) and dec==0:
            num=num*10+ord(i)-48
        if (ord(i)>=48 and ord(i)<=57) and dec==1:
            numdec=numdec*10+ord(i)-48
            c+=1
    if c==2:
        num=num+numdec*0.01
    if c==1:
        num=num+numdec*0.1
    return num

f=open("alumnos.txt","r")
l=f.read()
l=l.splitlines()
codigos=[]
cursos=[]
for i in l:
    linea=i.split()
    if linea[0] not in codigos:
        codigos.append(linea[0])
reg=[]
for i in range(len(codigos)):
    ap=0
    des=0
    for j in l:
        linea=j.split()
        if linea[0]==codigos[i]:
            if chartonum(linea[2])>=10:
                ap+=1
            else:
                des+=1
    reg.append([codigos[i],ap,des])
for i in range(len(reg)):
    print(f'Codigo de alumno es {reg[i][0]} : ')
    print(f'Cursos aprobados es : {reg[i][1]} ')
    print(f'Cursos desaprobados ess : {reg[i][2]} ')


for i in l:
    linea=i.split()
    if linea[1] not in cursos:
        cursos.append(linea[1])
max=0
for i in range(len(cursos)):
    ap=0
    des=0
    for j in l:
        linea=j.split()
        if linea[1]==cursos[i]:
            if chartonum(linea[2])>=10:
                ap+=1
            else:
                des+=1
    if(max<des):
        max=des
        mayordes=cursos[i]
print("La asignatura con mayoria en desaprobados viene a ser:",mayordes)
```
