```python
n=int(input("digite cuantos alumnos existen: "))
codigos=[]
nombres=[]
notas=[]
for i in range(n):
    cod=str(input("digite el código del alumno numero "+str(i+1)+" : "))
    while(len(cod)!=5):
        cod=str(input("digite un código de alumno válido (5 caracteres): "))
    codigos.append(cod)

    nombre=str(input("digite el nombre del alumno numero "+str(i+1)+" : "))
    while(len(nombre)>50):
        nombre=str(input("Nombre extenso, ingrese de nuevo: "))
    nombres.append(nombre)

    nota=float(input("Ingrese la nota del alumno numero "+str(i+1)+" : "))
    while(nota > 20 or nota < 0):
        nota = float(input("Nota no puede ser mayor que 20, ingrese de nuevo"))
    notas.append(nota)

x=[]


for i in range(n):
    x.append([0]*3)
for j in range(3):
    for i in range(n):
        if(j==0):
            x[i][j]=codigos[i]
        if(j==1):
            x[i][j]=nombres[i]
        if(j==2):
            x[i][j]=notas[i]
for i in range(n):
    for j in range(3):
        print(x[i][j],end=" ")
    print()

for i in range(1,len(notas)):
    for j in range(len(notas)-i):
        if(notas[j]<notas[j+1]):
            aux=notas[j]
            notas[j]=notas[j+1]
            notas[j+1]=aux
            aux1=codigos[j]
            codigos[j]=codigos[j+1]
            codigos[j+1]=aux1
            aux2=nombres[j]
            nombres[j]=nombres[j+1]
            nombres[j+1]=aux2


for j in range(3):
    for i in range(n):
        if(j==0):
            x[i][j]=codigos[i]
        if(j==1):
            x[i][j]=nombres[i]
        if(j==2):
            x[i][j]=notas[i]


print("*************")

for i in range(n):
    for j in range(3):
        print(x[i][j],end=" ")
    print()
```
