```python
import string

n=int(input("digite cuantas cadenas desea "))
cadenas = []
for i in range(n):
    cadena = str(input("Ingrese cadena posicion"+ str(i+1) + ": "))
    while(len(cadena)>50):
        cadena=str(input("error cadena muy grande vuelva a digitar"))
    cadenas.append(cadena)
abc = string.ascii_lowercase
filtrado = ""
for i in range(len(cadenas)):
    for c in cadenas[i]:
        if c in abc:
            filtrado += c
for c in abc:
    conteo = filtrado.count(c)
    if conteo > 0:
        print(c,end=" ")
        print(conteo, end=" ")
        print("*"*conteo)
```
