```python
def chartonum(l):
    dec=0
    num=0
    numdec=0
    c=0
    for i in l:
        if i==".":
            dec=1
        if (ord(i)>=48 and ord(i)<=57) and dec==0:
            num=num*10+ord(i)-48
        if (ord(i)>=48 and ord(i)<=57) and dec==1:
            numdec=numdec*10+ord(i)-48
            c+=1
    if c==2:
        num=num+numdec*0.01
    if c==1:
        num=num+numdec*0.1
    return num

f=open("VENTAS.txt","r+")
l=f.read()
l=l.splitlines()
montos=[]
nombres=[]
j=0
for i in l:
    linea=i.split()
    montos.append(chartonum(linea[0]))
    nombres.append(linea[1]+" "+linea[2])

for i in range(1,len(montos)):
    for j in range(len(montos)-i):
        if(montos[j]<montos[j+1]):
            (montos[j],montos[j+1])=(montos[j+1],montos[j])
            (nombres[j],nombres[j+1])=(nombres[j+1],nombres[j])

pos=[1]
aux=1
for i in range(1,len(montos)):
    if montos[i-1]>montos[i]:
        aux+=1
    pos.append(aux)

c=0
for i in range(len(pos)):
    if pos[i]<=5:
        c+=1
bonos=[]
porcentajes=[]
for i in range(len(montos)):
    if c!=0:
        por=11-pos[i]
        porcentajes.append(por)
        bono=(0.01)*(11-pos[i])*montos[i]
        bonos.append(bono)
        c-=1
    if c==0:
        porcentajes.append(11-(pos[i]+1))
        bonos.append(15023.7*(11-(pos[i]+1))/100)

for i in range(len(montos)):
    print(f"{round(float(montos[i]),2)}\t\t{nombres[i]}\t\t{pos[i]}\t{round(bonos[i],2)}\t\t{porcentajes[i]}%")
```

```python
n=int(input("Num vendedores: "))
names,notas=list([0]*n),list([0]*n)
for i in range(n):
  name=input("Ingrese nombre: ")
  nota=input("Ingrese venta: ")
  names[i],notas[i]=name,nota

for i in range(1,n):
  for j in range(n-i):
    if(notas[j]<notas[j+1]):
        (notas[j],notas[j+1])=(notas[j+1],notas[j])
index,c,lista=[1],1,[]
for i in range(1,n):
  if(notas[i]<notas[i-1]):
    c+=1
    index.append(c)
  elif notas[i]==notas[i-1]:
    index.append(c)
for i in range(n):
  lista.append([names[i],notas[i],index[i]])
print("Presentando datos...")
for i in range(n):
  print(lista[i][0],lista[i][1],lista[i][2])
    
```
