```python
n=int(input("Ingrese la cantidad de alumnos: "))
while n>30:
    n=int(input("Ingrese el numero de alumnos entre 0 y 30, que contega a 30 pero no a 0: "))
codigos=[]
pesos=[]
pesos2=[]
tallas=[]
for i in range(n):
    codigo=int(input(f"Ingrese el codigo del alumno #{i+1}: "))
    while not(10<=codigo<=99):
        codigo=int(input("Ingrese el codigo del alumno que sea de dos dígitos: "))
    codigos.append(codigo)
    peso=float(input(f"Ingrese el peso del alumno #{i+1}: "))
    pesos.append(peso)
    pesos2.append(peso)
    talla=float(input(f"Ingrese la talla del alumno #{i+1}: "))
    tallas.append(talla)

matriz=[]

for i in range(n):
    matriz.append([0]*3)
for j in range(3):
    for i in range(n):
        if(j==0):
            matriz[i][j]=codigos[i]
        if(j==1):
            matriz[i][j]=pesos[i]
        if(j==2):
            matriz[i][j]=tallas[i]
for i in range(n):
    for j in range(3):
        print(matriz[i][j],end=" ")


for i in range(1,len(pesos2)):
        for j in range(0,len(pesos2)-i):
            if(pesos2[j+1] < pesos2[j]):
                aux=pesos2[j]
                pesos2[j]=pesos2[j+1]
                pesos2[j+1]=aux
h=0
for i in pesos2:
  if i==matriz[h][1]:
    print(matriz[h])
    h=h+1
```
