```python
numero=int(input("Ingrese un numero: "))
while(numero<0):
    numero=int(input("Ingrese un numero positivo: "))
aux=numero
num=0
aux1=0
for i in range(2,10):
    while(aux>0):
        residuo=aux%i
        num=num*10+residuo
        aux=aux//i
    aux3=num

    while(aux3>0):
        residuo1=aux3%10
        aux1=aux1*10+residuo1
        aux3=aux3//10

    if(aux1==num and aux1>=10):
        print(f"El número es capicúa en la base: {i} (pues el número es {aux1})")
    aux=numero
    num=0
    aux1=0
```
