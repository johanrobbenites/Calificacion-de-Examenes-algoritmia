```python
filas=int(input("Ingrese el numero de filas: "))
columnas=int(input("Ingrese el numero de columnas: "))
m1=[]
for i in range(filas):
  m1.append([])
  for j in range(columnas):
    valor= int(input(f"Fila {i+1}, Columna {j+1} : "))
    m1[i].append(valor)

m_girada1 = []
m_girada2 = []
m_girada3 = []
m_girada4 = []

print("\nINICIAL:\n")
for i in range(len(m1)):
  print(m1[i])

for i in range(len(m1[0])):
    fila = []
    for j in range(len(m1) - 1, -1, -1):
        fila.append(m1[j][i])
    m_girada1.append(fila)
print("\nPrimera Vuelta:\n")
for i in range(len(m_girada1)):
  print(m_girada1[i])
  if i == len(m_girada1)-1:
    print("\nSegunda Vuelta:\n")

m2=m_girada1
for i in range(len(m2[0])):
    fila = []
    for j in range(len(m2) - 1, -1, -1):
        fila.append(m2[j][i])
    m_girada2.append(fila)
for i in range(len(m_girada2)):
  print(m_girada2[i])
  if i == len(m_girada2)-1:
    print("\nTercera Vuelta:\n")

m3=m_girada2
for i in range(len(m3[0])):
    fila = []
    for j in range(len(m3) - 1, -1, -1):
        fila.append(m3[j][i])
    m_girada3.append(fila)
for i in range(len(m_girada3)):
  print(m_girada3[i])
  if i == len(m_girada3)-1:
    print("\nCuarta Vuelta:\n")

m4=m_girada3
for i in range(len(m4[0])):
    fila = []
    for j in range(len(m4) - 1, -1, -1):
        fila.append(m4[j][i])
    m_girada4.append(fila)
for i in range(len(m_girada4)):
  print(m_girada4[i])
```

```python
base=int(input('Ingrese número en base 10: '))
for n in range(2,10):
    D=base
    q=D//n
    numero_str=''
    while q>=n:
        r=D%n
        numero_str=str(r)+numero_str
        D=q
        q=D//n
    numero_str=str(q)+numero_str
    if numero_str==numero_str[::-1]:
        print('El número es capicua en la.base:',n,'(pues el número es',numero_str,')')
```
