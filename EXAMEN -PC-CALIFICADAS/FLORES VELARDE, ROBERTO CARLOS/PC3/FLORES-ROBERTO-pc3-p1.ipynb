```python
n=int(input("Num estudiantes: "))
cods,names,notas=list([0]*n),list([0]*n),list([0]*n)
for i in range(n):
  cod=input("Ingrese codigo: ")
  name=input("Ingrese nombre: ")
  nota=input("Ingrese nota: ")
  cods[i],names[i],notas[i]=cod,name,nota
#sorting grades
for i in range(1,n):
  for j in range(n-i):
    if(notas[j]<notas[j+1]):
        (notas[j],notas[j+1])=(notas[j+1],notas[j])
index,c,lista=[1],1,[]
for i in range(1,n):
  if(notas[i]<notas[i-1]):
    c+=1
    index.append(c)
  elif notas[i]==notas[i-1]:
    index.append(c)
for i in range(n):
  lista.append([cods[i],names[i],notas[i],index[i]])
print("Presentando datos...")
for i in range(n):
  print(lista[i][0],lista[i][1],lista[i][2],lista[i][3])
```

```python
n=int(input("Cantidad de postulantes: "))
codigos=[]
nombres=[]
notas=[]
for i in range(n):
    codigo=input(f"codigo del postulante numero #{i+1} : ")
    while(len(codigo)!=5):
        codigo=input("codigo de alumno debe ser de 5 caracteres: ")
    codigos.append(codigo)

    nombre=input(f"nombre del postulante numero #{i+1} : ")
    while(len(nombre)>50):
        nombre=input("sobrepaso los 50 caracteres permitidos, ingrese de nuevo: ")
    nombres.append(nombre)

    nota=float(input(f"nota del postulante numero #{i+1} : "))
    while(nota > 20 or nota < 0):
        nota = float(input("Esta nota no esta en el rango de 0 a 20, ingrese de nuevo: "))
    notas.append(nota)

arreglo=[]

for i in range(n):
    arreglo.append([0]*3)
for j in range(3):
    for i in range(n):
        if(j==0):
            arreglo[i][j]=codigos[i]
        if(j==1):
            arreglo[i][j]=nombres[i]
        if(j==2):
            arreglo[i][j]=notas[i]
for i in range(1,len(notas)):
    for j in range(len(notas)-i):
        if(notas[j]<notas[j+1]):
            aux=notas[j]
            notas[j]=notas[j+1]
            notas[j+1]=aux
            aux1=codigos[j]
            codigos[j]=codigos[j+1]
            codigos[j+1]=aux1
            aux2=nombres[j]
            nombres[j]=nombres[j+1]
            nombres[j+1]=aux2
            
for j in range(3):
    for i in range(n):
        if(j==0):
            arreglo[i][j]=codigos[i]
        if(j==1):
            arreglo[i][j]=nombres[i]
        if(j==2):
            arreglo[i][j]=notas[i]


for i in range(n):
    for j in range(3):
        print(arreglo[i][j],end="  ")
    print()
    
```
