```python
def histograma():
    n=int(input("Cantidad de cadenas: "))
    cadenas = []
    for i in range(n):
        cadena = input(f"Ingrese cadena nmero #{i+1} : ")
        while(len(cadena)>50):
            cadena=input("Cadena muy larga, ingrese de nuevo: ")
        cadenas.append(cadena)

    abecedario = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    nuevo= ""
    for i in range(len(cadenas)):
        for j in cadenas[i]:
            if j in abecedario:
                nuevo += j

    for i in abecedario:
        contador = nuevo.count(i)
        if contador > 0:
            frec=contador*"*"
            print(f"{i}  {contador}  {frec}")
            frec=""
histograma()
```
