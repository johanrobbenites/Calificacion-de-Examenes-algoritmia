```python
def char_a_num(linea):
    aux3=0
    numero=0
    aux4=0
    k=0
    for i in linea:
        if i==",":
            aux3=1
        if (ord(i)>=48 and ord(i)<=57) and aux3==0:
            numero=numero*10+ord(i)-48
        if (ord(i)>=48 and ord(i)<=57) and aux3==1:
            aux4=aux4*10+ord(i)-48
            k+=1
    if k==2:
        numero=numero+aux4*0.01
    if k==1:
        numero=numero+aux4*0.1
    return numero

f=open("ALUMNOS.txt","r")
linea=f.read()
linea=linea.splitlines()
codigos=[]
cursos=[]
for i in linea:
    linea_separada=i.split()
    if linea_separada[0] not in codigos:
        codigos.append(linea_separada[0])
registro=[]
for i in range(len(codigos)):
    aux=0
    aux1=0
    for j in linea:
        linea_separada=j.split()
        if linea_separada[0]==codigos[i]:
            if char_a_num(linea_separada[2])>=10:
                aux+=1
            else:
                aux1+=1
    registro.append([codigos[i],aux,aux1])
for i in range(len(registro)):
    print(f'Codigo de alumno {registro[i][0]}... ')
    print(f'Cursos aprobados: {registro[i][1]} ')
    print(f'Cursos desaprobados: {registro[i][2]} ')


for i in linea:
    linea_separada=i.split()
    if linea_separada[1] not in cursos:
        cursos.append(linea_separada[1])
max=0
for i in range(len(cursos)):
    aux=0
    aux1=0
    for j in linea:
        linea_separada=j.split()
        if linea_separada[1]==cursos[i]:
            if char_a_num(linea_separada[2])>=10:
                aux+=1
            else:
                aux1+=1
    if(max<aux1):
        max=aux1
        cant_desaprobados=cursos[i]
print("El curso con mayor desaprobados es",cant_desaprobados)
```
