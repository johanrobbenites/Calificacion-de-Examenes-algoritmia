```python
def posicionElemento(lista,elementoBuscar, indice):
    
    if indice==len(lista) or lista[indice]==elementoBuscar:
        if indice==len(lista):
            posicion=-1
        else:
            posicion=indice
    else:
        posicion=posicionElemento(lista, elementoBuscar,indice +1)
    return posicion    
    
n=int(input("Ingrese la cantidad de numeros: "))
x=[]
for i in range(n):
    num=int(input(f'Ingrese el valor #{i+1} : '))
    x.append(num)
valor=int(input("Ingrese el numero a buscar: "))
print(posicionElemento(x,valor,0))
```

```python
def posicion(x, valor, inicio, fin):
    flag=True
    medio = (fin-inicio) // 2
    if x[medio] == valor:
        return medio
    if fin - inicio == 1:
        return -1
    if x[medio] < valor:
        return posicion(x, valor, inicio, medio)
    else:
        return posicion(x, valor, medio, fin)

n=int(input("Ingrese la cantidad de numeros: "))
x=[]
for i in range(n):
    num=int(input(f'Ingrese el valor #{i+1} : '))
    x.append(num)
valor=int(input("Ingrese el numero a buscar: "))
print(posicion(x,valor,0,n))
```
