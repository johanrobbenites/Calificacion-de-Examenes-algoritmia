```python
n=int(input("INGRESE LA CANTIDAD DE POSTULANTES: "))
datos=[]
for i in range(n):
  datos.append([0]*3)
for i in range(len(datos)):
  for j in range(3):
    if j==0:
      while True:
        codigo=input(f"INGRESE CODIGO DEL ALUMNO {i+1}: ")
        if len(codigo)==5:
          break
      datos[i][j]=codigo
    if j==1:
      while True:
        nombre=input(f"INGRESE NOMBRE DEL ALUMNO {i+1}: ")
        if len(nombre)<=50:
          break
      datos[i][j]=nombre
    if j==2:
      while True:
        nota=float(input(f"INGRESE NOTA DEL ALUMNO {i+1}: "))
        m=str(nota)
        if int(nota)<=20 and len(m)<=6:
          break
      datos[i][j]=nota      
for i in range(1,len(datos)):
  for j in range(len(datos)-i):
    if datos[j][2]<datos[j+1][2]:
       m=datos[j][2]
       datos[j][2]=datos[j+1][2]
       datos[j+1][2]=m
for i in range(len(datos)):
  print(f"{datos[i][0]} {datos[i][1]} {datos[i][2]} {i+1}")



  

  






```
