```python
n=int(input("INGRESE LA CANTIDAD DE CADENAS: "))
lista=[]
for i in range(n):
  while True:
     texto=input(f"INGRESE LA CADENA {i+1}: ")
     if len(texto)<=50:
       break
  lista.append(texto)
m=" ".join(lista)
for i in range(97,123):
  k=m.count(chr(i))
  if k!=0:
    print(chr(i) ,k,"*"*k)


```
