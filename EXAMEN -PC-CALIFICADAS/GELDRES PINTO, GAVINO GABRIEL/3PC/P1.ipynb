```python
n = int(input("INGRESE LA CANTIDAD DE POSTULANTES: "))
datos = {}
notas = {}
for i in range(n):
  codigo = input("INGRESE EL CODIGO DE POSTULANTE: ")
  nombre = (input("INGRESE EL NOMBRE DEL POSTULANTE: "))
  nota = float(input("INGRESE LA NOTA: "))
  datos[nombre] = codigo
  notas[nombre] = nota

mayor = 0
for valor in notas.values():
  if valor > mayor:
    mayor = valor

def get_key(val):
  for key, value in notas.items():
    if val == value:
      return key
  return get_key(mayor)
mayorNota = get_key(mayor)

def get_key(val):
  for key, value in datos.items():
    if val == key:
      return value
  return get_key(get_key(mayor))
mayorCodigo = get_key(mayorNota)

print("CODIGO\t\tNOMBRE\t\tPUNTAJE")
print(f"{mayorCodigo}\t\t{mayorNota}\t\t{mayor}")
```
