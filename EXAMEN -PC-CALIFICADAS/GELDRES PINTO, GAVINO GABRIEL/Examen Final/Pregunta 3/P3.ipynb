```python
def palindromo_recursivo(cadena, i, f):
    if i >= f:
        return "es palíndroma"
    if cadena[i] == cadena[f]:
        return palindromo_recursivo(cadena, i+1, f-1)
    else:
        return "no es palíndroma"



def acomodar(cadena):
    cadena = cadena.lower()
    cadena = cadena.replace(" ", "")
    return cadena

def invertir(x):
  return x [::-1]

texto = str(input("Ingrese la cadena: "))
palabras = [texto]
for palabra in palabras:
    palabra_limpia = acomodar(palabra)
    es_palindromo_recursivo = palindromo_recursivo(
        palabra_limpia, 0, len(palabra_limpia)-1)
    print(f"La frase invertida es: {invertir(acomodar(palabra))} por lo tanto {es_palindromo_recursivo}")
```
