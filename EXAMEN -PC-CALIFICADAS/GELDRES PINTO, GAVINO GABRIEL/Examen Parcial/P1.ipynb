```python

m = []
for i in range(5):
  m.append([0]*3)
print(m)
codigo = -18
while codigo != 0:
  codigo_est = int(input("INGRESE EL CODIGO DEL ESTUDIANTE: "))
  peso = int(input("INGRESE EL PESO: "))
  talla = int(input("INGRESE LA TALLA: "))
  if (codigo >= 10 and codigo <= 99) and talla > 0 and peso > 0:
    m[codigo-1][codigo_est-1] = m[codigo-1][codigo_est-1] + peso

caract = ["PESO", "TALLA"]
for i in caract:
  print(i)
  for j in range(3):
    print("CODIGO DE ESTUDIANTE: ", j+1, " ", m[i][j])
```
