Escriba un programa que lea n cadenas de tamaño no mayor a 50 caracteres, y realice un histograma con las frecuencias del caracter ‘a’ hasta el caracter ‘z’. No se consideran las vocales tildadas, ni la ‘ñ’. Ejemplo, si n = 4 y se ingresan las siguientes cadenas:

```python
diccionario = {'a':0,'b':0,'c':0,'d':0,'e':0,'f':0,'g':0,'h':0,'i':0,'j':0,'k':0,
               'l':0,'m':0,'n':0,'o':0,'p':0,'q':0,'r':0,'s':0,'t':0,'u':0,'v':0,'w':0,'x':0,'y':0,'z':0}
lcadenas = []
n = int(input("Numero de cadenas: "))
for c in range(n):
  cadena = str(input(f"Cadena {c+1}: "))
  lcadenas.append(cadena)
cdefinitivo = "".join(lcadenas)

for d in cdefinitivo:
  if d in diccionario.keys():
    if diccionario[d]==0:
      diccionario[d] = 1
    else:
      diccionario[d] += 1
#print(diccionario)
# {'a': 5, 'b': 2, 'c': 1, 'd': 0, 'e': 1, 'f': 0, 'g': 0, 'h': 1, 'i': 0, 'j': 0, 'k': 0, 'l': 1, 'm': 1, 
#  'n': 2, 'o': 6, 'p': 1, 'q': 0, 'r': 3, 's': 2, 't': 2, 'u': 2, 'v': 0, 'w': 0, 'x': 0, 'y': 2, 'z': 0}
for i in diccionario.keys():
  if diccionario[i]>0:
    print(f"{i}  {diccionario[i]} {'*'*diccionario[i]}")

#Hola María ¿cómo estás?
#Juan Mora
#Python o Ruby?
#Aprobé SI205
```
