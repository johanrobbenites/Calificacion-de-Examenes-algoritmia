Escriba un programa que lea n números enteros y los almacene en un arreglo, luego solicite un número y mediante una función recursiva busque el número en el arreglo devolviendo la posición en que aparece por primera vez. Si el número no está en el arreglo debe devolver -1.

```python
def busqueda(arreglo, x):
  global indice
  indice+=1
  if x not in arreglo:
    return -1
  if arreglo[0]==x:
    return indice
  else:
    return busqueda(arreglo[1:],x)

enteros = []
n = int(input("Ingrese n: "))
for i in range(n):
  x=int(input("Ingrese número: "))
  enteros.append(x)

b = int(input("Número a buscar: "))
indice = -1

print(enteros)
print(busqueda(enteros, b))  

```
