```python
seccciones={"U":[],"V":[],"W":[]}
with open("/content/programa.txt","r",encoding="utf-8") as al:
    for line in al:
        li=line.split(" ")
        for i in li:
            if i[0] == "U":
                seccciones["U"].append(int(i[1:]))
            if i[0] == "V":
                seccciones["V"].append(int(i[1:]))
            if i[0] == "W":
                seccciones["W"].append(int(i[1:]))

for i in range(len([key for key in seccciones.keys()])):
    llave=[key for key in seccciones.keys()][i]
    prom=0
    sum=0
    for o in seccciones[llave]:
        prom=prom+1
        sum=sum+o
    seccciones[llave].append(sum/prom)

mayor=0
for i in range(len([key for key in seccciones.keys()])):
    llave=[key for key in seccciones.keys()][i]
    if seccciones[llave][-1]>mayor:
        mayor=seccciones[llave][-1]
        
for i in range(len([key for key in seccciones.keys()])):
    llave=[key for key in seccciones.keys()][i]
    if seccciones[llave][-1]==mayor:
        print(f"Seccion {llave} promedio {seccciones[llave][-1]}")
```
