```python
ventas={}
with open("ventas.txt","r",encoding="utf-8") as ve:
    for line in ve:
        li=line.split(" ",1)
        if li[0] in [key for key in ventas.keys()]:
            ventas[float(li[0])].append(li[1][:-1])
        else:
            ventas[float(li[0])]=[li[1][:-1]]
#ordenando de mayor a menor:
llavesOrdenadas=[]
for i in range(len([key for key in ventas.keys()])):
    llave=[key for key in ventas.keys()][i]
    llavesOrdenadas.append(llave)
for i in range(1,len(llavesOrdenadas)):
        for j in range(0,len(llavesOrdenadas)-i):
            if(llavesOrdenadas[j+1] < llavesOrdenadas[j]):
                aux=llavesOrdenadas[j]
                llavesOrdenadas[j]=llavesOrdenadas[j+1]
                llavesOrdenadas[j+1]=aux

#Imprimiendo:
print("MONTO \t Nombre \t Pto \t BONO \t Porc")
numero=0
pto=1
for i in llavesOrdenadas:
    bono=11-pto
    for o in ventas[i]:
        numero=numero+1
        if numero>5:
            exit()
        else:
            print(f"{i}\t{o}\t{pto}\t{(i*bono)/100}\t{bono}%")
    pto=pto+len(ventas[i])
```
