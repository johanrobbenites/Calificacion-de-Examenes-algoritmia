```python
oracion=input("Ingrese la oracion: ")
oracion=oracion.replace(" ","")
oracion=oracion.lower()
print(oracion)

def pali(x):
    if x=="":
        inv=x
    else:
        inv= x[-1]+pali(x[:-1])
    return inv
a= pali(oracion)
if a==oracion:
    print("Es palindromo")
else:
    print("No es palindromo")
    
```
