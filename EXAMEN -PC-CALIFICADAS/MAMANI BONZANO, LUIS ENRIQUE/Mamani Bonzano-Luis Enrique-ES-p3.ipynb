```python
def contar(pal, ora):
  ora = ora.split()
  return ora.count(pal)

ora = input("Ingrese una oracion: ")
pal = input("Ingrese una palabra: ")
print(contar(pal, ora))
```

```python
def contar(pal, ora):
  contador = 0
  for i in range(len(ora)):
    if ora[i] == pal[0] and ora[i+1] == pal[1]:
      contador += 1
  return contador

ora = input("Ingrese una oracion: ")
pal = input("Ingrese una palabra: ")
print(contar(pal, ora))
```
