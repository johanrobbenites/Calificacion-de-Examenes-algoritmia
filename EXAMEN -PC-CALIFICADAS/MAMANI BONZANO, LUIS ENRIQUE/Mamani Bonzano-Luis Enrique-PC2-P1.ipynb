PREGUNTA 1: Escriba un programa, que lea las notas de los n alumnos que dieron un examen y determine el porcentaje de aprobados. Si el porcentaje es inferior a 70% debe añadir un punto a cada examen y recalcular el porcentaje de aprobados y así sucesivamente hasta igual o superar 70%. Considere que cuando un examen llega a 20 ya no tiene que añadir puntos. Su programa debe mostrar el porcentaje de aprobados de cada iteración.

```python
n = int(input("Ingrese la cantidad de alumnos: "))
notas = [0]*n
apro = 0

for i in range(n):
  notas[i] = int(input("Ingrese las notas: "))
  a = notas[i]
  if a >= 10:
    apro = apro + 1
cant = (apro/n)*100
print(notas,",", cant,"% de aprobados")

while cant < 70:
    apro = 0
    for j in range(n):
      notas[j] = notas[j] + 1
      a = notas[j]
      if a >= 10:
        apro = apro + 1
    cant = (apro/n)*100
    print(notas,",", cant,"% de aprobados")
```
