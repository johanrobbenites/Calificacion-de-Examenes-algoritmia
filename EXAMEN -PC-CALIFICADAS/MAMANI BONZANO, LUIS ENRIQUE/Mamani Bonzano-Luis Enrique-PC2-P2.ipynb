PREGUNTA 2: Una tienda tiene 4 empleados y 3 productos. Se desea conocer al final del día cuanto de cada producto ha vendido cada uno de los empleados, mostrando los resultados en forma ordenada por empleado y cantidad vendida de producto de menor a mayor. Los datos se ingresan por cada venta considerando que el código del empleado (del 1 al 4), el código del producto (del 1 al 3) y la cantidad vendida. Una venta solo involucra un producto. Para terminar el ingreso de datos se digita 0 como código del empleado.

```python
v = []
for i in range(4):
  v.append([0]*3)
print(v)

code=int(input("Ingrese código del empleado [1 al 4]: "))
while code != 0:
  codp=int(input("Ingrese el codigo del producto [1 al 3]: "))
  cant=int(input("Ingrese la cantidad: "))
  if (code>=1 and code<=4) and (codp>=1 and codp<=3) and cant>0:
    v[code-1][codp-1]=v[code-1][codp-1]+cant
  code=int(input("Ingrese código del empleado [1 al 4]: "))

mayor = 0
menor = 1000
for i in range (4):
  print("Empleado ", i+1,":")
  for j in range (3):
    if menor >= v[i][j]:
      menor = v[i][j]
    print("Codigo del producto: ",j+1," ",menor)



print(v)
```
