Pregunta 3: Diseñe un programa que permita almacenar en dos vectores de enteros los dígitos de dos números en el sistema de numeración de base b. Luego, sin convertir a decimal, realizar la multiplicación los números y almacenar en otro vector los dígitos del producto de dichos números.

```python
b = int(input("Introduzca la base del sistema: "))

c1= int(input("Cantidad de dígitos del primer número: "))
v1 = [0]*c1
for i in range(c1):
  m = int(input("Ingrese digito:"))
  while m >= b:
     m = int(input("Ingrese digito:"))
  if m < b:
    v1[i] = m  
print(v1)

c2= int(input("Cantidad de dígitos del segundo número: "))
v2 = [0]*c1
for i in range(c2):
  n = int(input("Ingrese digito:"))
  while n >= b:
     n = int(input("Ingrese digito:"))
  if n < b:
    v2[i] = n
print(v2)

c = c1 + c2
v2 = [0]*c
```
