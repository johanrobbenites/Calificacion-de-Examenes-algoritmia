```python
#
b =[]
n = int(input("Ingrese el numero de filas de la matriz:"))
m = int(input("Ingrese el numero de columnas de la matriz"))
for i in range(n):
  b.append([0]*m)
for i in range(n):
  for j in range(m):
    dato = int(input("Ingrese el dato -----> "))
    b[i][j] = dato
print("\n", b)    
```
