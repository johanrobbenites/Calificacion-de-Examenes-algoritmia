```python
"""
Crear una función que permita recibir en dos cadenas los conjuntos de valores 
enteros A y B, luego en otras dos cadenas devuelva la unión e intersección de
dichos conjuntos.
"""
un1 = []
in1 = []
m = int(input("Ingrese la cantidad de elementos: "))
a = [0] * m
for i in range(m):
  a[i] = int(input("Ingrese los datos: "))
print("CONJUNTO A")
print(a)

n = int(input("Ingrese la cantidad de elementos: "))
b = [0] * n
for i in range(n):
  b[i] = int(input("Ingrese los datos: "))
print("CONJUNTO B")
print(b)

def union(x, y):
    lista_union = sorted(x + y)
    return lista_union
print("UNIÓN =", union(a,b))

for element in b:
    if element in a:
        in1.append(element)
print("INTERSECCIÓN =", in1)
```
