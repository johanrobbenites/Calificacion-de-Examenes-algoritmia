```python
# Escriba un programa que lea n numeros enteros y los almacene en un arreglo. 
# Luego solicite un numero y mediante una funcion recursiva busque el numero en el arreglo devolviendo la posicion en que aparece por primera vez. 
# Si el numero no está en el arreglo debe devolver -1.
n = int(input("NUMERO DE ELEMENTOS DEL ARREGLO: "))
a = [0] * n
ix = [0] * n
for i in range(n):
  a[i] = int(input("INGRESE LOS NUMEROS AL ARREGLO: "))
print(a)
num_sol = int(input("INGRESE EL NÚMERO QUE DESEA BUSCAR: "))
i = 0
while i < n:
  if a[i] == num_sol:
    j=i
    while j<n:
      a[j] = a[j+1]
      j = j+1
      n-=1
  else:
    i+=1
```
