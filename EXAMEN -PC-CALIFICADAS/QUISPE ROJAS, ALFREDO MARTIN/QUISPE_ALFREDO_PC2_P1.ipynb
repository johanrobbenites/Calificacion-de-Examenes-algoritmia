```python
lista = []
n = int(input("INGRESE LA CANTIDAD DE NOTAS =====> "))
for i in range(n):
  x = int(input("INGRESE LAS NOTAS: "))
  lista.append(x)

lista_no_redundante = []
for x in lista:
  if x >= 10:
    lista_no_redundante.append(x)

for x in lista_no_redundante:
  promedio = sum(x) / n
print(f"{promedio}% DE APROBADOS")

```
