```python
num1 = int(input("INGRESE UN NÚMERO DE TRES CIFRAS: "))
b = int(input("INGRESE LA BASE: "))
if num1 >= 100 and num1 <= 999:
  n1 = num1 % b
  n2 = (num1 // b) % b
  n3 = (num1 // b) // b 
  nuevo1 = (n3 * 100) + (n2 * 10) + n1
  print(f"EL NÚMERO A BASE {b} ES {nuevo1}")
  list = [n3 , n2, n1]
  print(f"LOS DÍGITOS DEL PRIMER NÚMERO EN BASE {b}: ", list)
else:
  print("EL NÚMERO NO CONTIENE TRES CIFRAS")
num2 = int(input("INGRESE UN NÚMERO DE CUATRO CIFRAS: "))
if num2 >= 1000 and num2 <= 9999:
  m1 = num2 % b
  m2 = (num2 // b) % b
  m3 = ((num2 // b) // b) % b
  m4 = ((num2 // b) // b) // b
  nuevo2 = (m4 * 1000) + (m3 * 100) + (m2 * 10) + m1
  print(f"EL NÚMERO A BASE {b} ES {nuevo2}")
  list2 = [m4 , m3, m2, m1] 
  print(f"LOS DÍGITOS DEL SEGUNDO NÚMERO EN BASE {b}: ", list2)
else:
  print(f"EL NÚMERO NO CONTIENE CUATRO CIFRAS")
  

```
