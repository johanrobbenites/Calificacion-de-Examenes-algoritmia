```python
num = int(input("Ingrese el número: "))
n = int(input("Ingrese el número de cifras: "))
for i in range(1,10):
  while True:
    res = num % i
    num = num // i
    if num == 0:
      break
num = (num * 10) + res
print(f"El número en base {i} es: {num}")
if (num // 111) != 0:
  print(f"El número {num} es capicúa")
```
