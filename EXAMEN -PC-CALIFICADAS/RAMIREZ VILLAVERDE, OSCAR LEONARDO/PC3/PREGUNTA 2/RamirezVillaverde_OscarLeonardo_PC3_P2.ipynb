```python
n = int(input("Ingrese el valor de n cadenas:"))
while n > 50:
  print("!ERROR! El valor de n no debe exceder de 50")
  n = int(input("Ingrese el valor de n cadenas:"))
a = [0]*n
m = [0]*26
letras = ["a","b","c","d","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
for i in range(n):
    a[i]=input(f"Ingrese la cadena {i+1} : ")
for i in range(n):
    beta = a[i]
    for j in range(len(beta)):
        if (beta[j] >= "a") and (beta[j] <="z") and (beta[j]!="Ã±"):
            for x in range(len(letras)):
                if beta[j]==letras[x]:
                    m[x]= m[x]+1
for i in range(len(letras)):
    if m[i]!=0:
        print(f"{letras[i]}  {m[i]} ", "*"*m[i] ) 
```
