```python
n = int(input("ingrese el numero de postulantes : "))
cod = []
nom = []
punt = []
for i in range(n):
  c = input("ingrese el codigo : ")
  if len(c)== 5:
    cod.append(c)
for i in range(n):
  no = input("ingrese el nombre : ")
  if len(no) < 51:
    nom.append(no)
for i in range(n):
  p = float(input("ingrese el puntaje : "))
  punt.append(round(p,3))
for i in range(n) :
 print("el codigo :",cod[i],"-> su nombre :" ,nom[i],"-> el puntaje :",punt[i])
#ordenado los puntajes
for i in range(0,n):
  for j in range(i+1,n):
    if punt[i]<punt[j]:
      t = punt[i]
      punt[i]=punt[j]
      punt[j]=t
print("los puntajes ordenados :" , punt)
```
