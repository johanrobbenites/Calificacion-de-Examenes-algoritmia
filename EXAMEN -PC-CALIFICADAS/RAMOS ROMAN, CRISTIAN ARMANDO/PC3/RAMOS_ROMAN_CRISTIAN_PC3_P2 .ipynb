```python
n = int(input("ingrese la cantidad de frases: "))
def contcar(n):
 for i in range(n):
  frase = input("ingrese la frase : ")
  conteo = {}
  for letra in frase.lower():
   if letra not in conteo :
    conteo[letra]= 1
   else :
    conteo[letra] += 1
  for k , v in conteo.items():
   print("{} : {} ".format(k,v))
print(contcar(n))
```
