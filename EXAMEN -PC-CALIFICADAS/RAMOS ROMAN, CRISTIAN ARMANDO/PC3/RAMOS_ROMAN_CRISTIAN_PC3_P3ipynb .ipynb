```python
a = []
b = []
n1 = int(input("ingrese la cantidad de elementos de la cadena a : "))
n2 = int(input("ingrese la cantidad de elementos de la cadena b : "))
for i in range(n1):
  num = int(input("ingrese un numero entero :"))
  a.append(num)
A = set(a)
print("EL CONJUNTO  A = ",A)
for i in range(n2):
  num = int(input("ingrese un numero entero :"))
  b.append(num)
B = set(b)
print("EL CONJUNTO B =  " , B)
union = A | B
print("LA UNION = " , union)
interseccion = A & B
print("LA INTERSECCION = " , interseccion)
```
