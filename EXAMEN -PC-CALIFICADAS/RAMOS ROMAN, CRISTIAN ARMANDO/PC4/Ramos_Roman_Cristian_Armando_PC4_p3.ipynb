```python

n = int(input("ingrese un numero entero :"))
l = []
for i in range(n):
  x=int(input("ingrese un numero :"))
  l.append(x)
print(l)
num = int(input("ingrese el numero a solicitar : "))
def posicion(lista,n:int,x:int,i=0):
  if i < n-1 :
    if lista[i] == x :return i
    else :return posicion(lista,n,x,i+1)
  else :
    if lista[i]==x: return i
    else: return -1
p= posicion(l,n,num,)
print("la posicion del ",num,"es :",p)





    

  
```
