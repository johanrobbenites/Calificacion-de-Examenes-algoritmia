```python
def UandI(A,B):
  Union=[]
  Inter=[]
  for i in range(len(A)):
    if(A[i] in B):
      if ((A[i] in Inter)==False):
        Inter.append(A[i])
    if ((A[i] in Union)==False):
      Union.append(A[i])
  for j in range(len(B)):
    if ((B[j] in Union)==False):
      Union.append(B[j])
  return [Union,Inter]
A=input("A: ")
B=input("B: ")
A=A.split(sep=";")
B=B.split(sep=";")
C=UandI(A,B)
print("UNION",*C[0])
print("INTERSECCIÓN",*C[1])
print("FIN")
```
