```python
def inv_S_espacios(cadena,i):
  if(i==0):
    cadena2=""
    for line in cadena.split(" "):
      cadena2=cadena2+line
    cadena=cadena2
  cadena=list(cadena)
  aux=cadena[len(cadena)-1-i]
  cadena[len(cadena)-1-i]=cadena[i]
  cadena[i]=aux
  if i+1<len(cadena)-2-i:
    return inv_S_espacios(cadena,i+1)
  else:
    return cadena
def cadena_S_espacios(cadena):
  cadena2=""
  for line in cadena.split(" "):
    cadena2=cadena2+line
  cadena2=list(cadena2)
  return cadena2
cadena=input("Ingrese la cadena: ")
if inv_S_espacios(cadena,0)==cadena_S_espacios(cadena):
  print("Es palíndromo")
else :
  print("No es palíndromo")
```
