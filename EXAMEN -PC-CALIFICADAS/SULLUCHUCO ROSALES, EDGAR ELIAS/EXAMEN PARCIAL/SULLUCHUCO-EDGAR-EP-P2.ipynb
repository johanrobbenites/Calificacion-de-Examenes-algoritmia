```python
n=int(input("Ingrese el numero: "))
aux1=n
for b in range(2,10):
  nib=0
  n=aux1
  while(n>0):
    nib=n%b+nib*10
    n=n//b
  aux2=nib
  nb=0
  while(aux2>0):
    nb=nb*10+aux2%10
    aux2=aux2//10
  if(nib==nb):
    print(f"El número es capicua en base: {b} (pues el número es {nb})")
```
