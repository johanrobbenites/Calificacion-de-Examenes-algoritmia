```python
V=[]
for i in range(4):
  a=[]
  for j in range(4):
    a.append(0)
  V.append(a)
for i in range(4):
  #codigo del empleado
  V[i][3]=i+1
codE=int(input("código del empleado (1-4): "))
while(codE!=0):
  codP=int(input("código del producto (1-3): "))
  cantidad=int(input("cantidad del producto: "))
  V[codE-1][codP-1]=V[codE-1][codP-1]+cantidad
  codE=int(input("código del empleado (1-4): "))
c=[0,0,0,0]
for i in range(4):
  for j in range(3):
    c[i]=c[i]+V[i][j]
for i in range(4):
  for j in range(3):
    if(c[j]<c[j+1]):
      aux=c[j]
      c[j]=c[j+1]
      c[j+1]=aux
      aux=V[j]
      V[j]=V[j+1]
      V[j+1]=aux
print(*V)
```
