```python
b=int(input("base: "))
t=True
while(t):
  t=False
  n1=input(f"primer número (base {b}): ")+"f"
  i=0
  A=[]
  c1=0
  while(n1[i]!="f"):
    A.append(int(n1[i]))
    c1=c1+1
    if(int(n1[i])>=b):
      t=True
    i=i+1
print(*A)
t=True
while(t):
  t=False
  n2=input(f"segundo número (base {b}): ")+"f"
  i=0
  B=[]
  c2=0
  while(n2[i]!="f"):
    B.append(int(n2[i]))
    c2=c2+1
    if(int(n1[i])>=b):
      t=True
    i=i+1
print(*B)
C=[0]
x=0
for i in range(c1-1,0,-1):
  for j in range(c2-1,0,-1):
    C[x+(c1-i)-1]=C[x+(c1-i)-1]+i*j
    c=0
    while(C[x+(c1-i)-1]>=b):
      c=c+1
      if(c==1):
        C.append(0)
      C[x+(c1-i)-1]=C[x+(c1-i)-1]-b
      C[x+1+(c1-i)-1]=C[x+1+(c1-i)-1]+1
    if(c==0):
      C.append(0)
print(*C)

```
