```python
from google.colab import drive
drive.mount("/content/drive/")

arch=open("archivo.txt","w")
nom=input("ingrese los datos: (ya existe un archivo en drive, copie de ahi)")
arch.write(nom)
arch.close()

l1=open("archivo.txt","r")
v=l1.readlines()
l1.close()
cadena="".join(v)
print(cadena)
lista=cadena.split(", ")
orden=sorted(lista)
inv=orden[::-1]
for i in range(len(inv)-5):
    inv=inv[:-1]
print(inv)
extraer=" ".join(inv)
monto=extraer.split(" ")
b=[]
for i in range(int(len(monto)/3)):
    b.append(float(monto[3*i]))
for i in range(len(b)-5):
    b=b[:-1]
print(b)
print("--------------------------------------")
for i in range(len(inv)):
  print(inv[i],"con puesto",i+1,"recibe un bono de",float(b[i]*(10-i)/100),"del porcentaje",(10-i),"%")
```
