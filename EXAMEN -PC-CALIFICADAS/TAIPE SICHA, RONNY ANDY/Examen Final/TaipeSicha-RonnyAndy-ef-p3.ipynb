```python
cad=input("introduce una cadena :")
caracter=" "
for x in range(len(caracter)):
    cad = cad.replace(caracter[x],"")
palabra=cad.lower()
print(palabra)
def reversa(a):
  if len(a)==0:
    return a
  else:
    return reversa(a[1:])+a[0]
m=reversa(palabra)
print(m)
if m==palabra:
  print("la palabra es un palindromo")
else:
  print("la palabra no es un palindromo")
```
