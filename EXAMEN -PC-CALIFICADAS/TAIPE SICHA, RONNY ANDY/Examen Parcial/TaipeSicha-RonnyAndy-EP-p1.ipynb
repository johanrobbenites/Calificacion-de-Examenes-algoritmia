```python
matriz=[]
a=int(input("ingrese el numero de alumnos :"))
for i in range(a):
  matriz.append([0]*3)
for i in range(a):
  x=int(input("ingrese el codigo del alumno :"))
  y=float(input("ingrese el peso: "))
  z=float(input("ingrese la talla: "))
  matriz[i][0]=matriz[i][0]+x
  matriz[i][1]=matriz[i][1]+y
  matriz[i][2]=matriz[i][2]+z
for i in range(a):
  for j in range(a):
    if i<j:
      if matriz[i][1]<matriz[j][1]:
        l=matriz[i]
        matriz[i]=matriz[j]
        matriz[j]=l
for i in range(a):
  for j in range(a):
    if i<j:
      if matriz[i][1]==matriz[j][1]:
        if matriz[i][2]>matriz[j][2]:
          k=matriz[i]
          matriz[i]=matriz[j]
          matriz[j]=k
print("la lista de alumnos sin el nuevo ingresante ordenada será: ",matriz)
matriz.append([0]*3)
x=int(input("ingrese el codigo del alumno AGREGADO :"))
y=float(input("ingrese el peso del alumno AGREGADO: "))
z=float(input("ingrese la talla del alumno AGREGADO: "))
matriz[3][0]=matriz[3][0]+x
matriz[3][1]=matriz[3][1]+y
matriz[3][2]=matriz[3][2]+z
for i in range(a+1):
  for j in range(a+1):
    if i<j:
      if matriz[i][1]<matriz[j][1]:
        l=matriz[i]
        matriz[i]=matriz[j]
        matriz[j]=l
for i in range(a+1):
  for j in range(a+1):
    if i<j:
      if matriz[i][1]==matriz[j][1]:
        if matriz[i][2]>matriz[j][2]:
          k=matriz[i]
          matriz[i]=matriz[j]
          matriz[j]=k
print("la lista de alumnos con el nuevo ordenada será: ",matriz)
```
