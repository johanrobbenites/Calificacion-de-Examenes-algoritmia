```python
lista=[]
a=int(input("ingrese el numero de alumnos: "))
for i in range(a):
  x=int(input("ingrese la nota del estudiante: "))
  lista.append(x)
while True:
    c=0
    m=0
    for j in range(a):
      if lista[j]>=10:
        c=c+1
        m=c/a
    if m<0.7:
      for i in range(a):
        lista[i]=lista[i]+1
      for i in range(a):
        if lista[i]>=10:
            c=c+1
            m=c/a
    if m>0.7:
        break
for i in range(a):
    lista[i]=lista[i]+1
print(lista)
print("el porcentaje de aprobados es: ",m)
```
