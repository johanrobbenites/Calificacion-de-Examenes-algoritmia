```python
m=[]
for i in range(4):
    m.append([0]*3)
x=-1
while x!=0:
  x=int(input("ingrese el codigo del empleado :"))
  y=int(input("ingrese el codigo del producto :"))
  z=int(input("ingrese la cantidad del producto :"))
  if (x>=1 and x<=4) and (y>=1 and y<=3) and z>0:
    m[x-1][y-1]=m[x-1][y-1]+z
  print("------------------------------")
print(m)
for i in range(4):
  print("empleado",i+1)
  for j in range(3):
    print("producto",j+1, ":",m[i][j])
```
