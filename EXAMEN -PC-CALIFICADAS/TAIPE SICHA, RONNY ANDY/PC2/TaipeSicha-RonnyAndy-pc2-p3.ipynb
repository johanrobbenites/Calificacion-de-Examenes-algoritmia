```python
list1=[]
list2=[]
a=int(input("ingrese la base de la operación: "))
b=-1
while b<a:
   b=int(input("ingrese los digitos de el primer numero: "))
   if b==a:
     break
   if a>b>=0:
     list1.append(b)
c=-1
while c<a:
   c=int(input("ingrese los digitos de el segundo numero: "))
   if c==a:
     break
   if a>c>=0:
     list2.append(c)
print("los primeros numeros en base",a,"son :",list1)
print("los segundos numeros en base",a,"son :",list2)
m=0
k=len(list1)
for i in range(k):
  m=m+list1[i]*(a**(k-i-1))
z=len(list2)
h=0
for j in range(z):
  h=h+list2[j]*(a**(z-j-1))
multi=m*h
numerobase=[]
while multi>=1:
  numerobase.append(str(multi%a))
  multi//=a
print("".join(numerobase[::-1]))
```
