```python
import random
alumno={}
c=[]
n=int(input("ingrese el numero de estudiantes"))
for i in range(n):
  nombre=input("ingrese el nombre :")
  codigo=input("ingrese el codigo :")
  for i in range(n):
    b=random.uniform(0,20)
    a=round(b,3)
  listax=[nombre,a,0]
  c.append(listax)
  if codigo not in alumno:
    alumno[codigo]=listax
for i in range(n):
  for j in range(n):
    if i>j:
      if c[i][1]>c[j][1]:
        t=c[i][1]
        c[i][1]=c[j][1]
        c[j][1]=t
l=1
for i in alumno:
  m=list(alumno[i])
  m[2]=l
  l=l+1
  alumno[i]=m
print(alumno)
```
