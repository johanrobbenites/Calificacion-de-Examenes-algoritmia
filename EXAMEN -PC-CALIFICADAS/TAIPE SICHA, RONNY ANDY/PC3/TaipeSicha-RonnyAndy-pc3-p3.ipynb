```python
A=[]
B=[]
x=input("ingrese una cadena A con los valores separados por un (;)")
z=input("ingrese una cadena B con los valores separados por un (;)")
A=x.split(";")
B=z.split(";")
def union(a,b):
  C=[]
  for i in range(len(a)):
    if a[i] not in C:
      C.append(a[i])
  for i in range(len(b)):
    if b[i] not in C:
      C.append(b[i])
  return(C)

def inter(a,b):
  D=[]
  for i in range(len(a)):
    for j in range(len(b)):
      if a[i]==b[j]:
        D.append(a[i])
  return(D)
u=union(A,B)
j=inter(A,B)
print(";".join(map(str,u)))
print(";".join(map(str,j)))
```
