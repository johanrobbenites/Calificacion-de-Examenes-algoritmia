```python
cada=[]
while True:
  x=int(input("INTRODUCE NUMERO PARA A: "))
  cada.append(x)
  p=input("DESEA SEGUIR AGREGANDO NUMEROS S/N: ")
  if p=='N' or p=='n':
    break;

cadb=[]
while True:
  y=int(input("INTRODUCE NUERMO PARA B: "))
  cadb.append(y)
  m=input("DESEA SEGUIR AGREGANDO NUMEROS S/N: ")
  if m=='N' or m=='n':
    break;

def union(cada,cadb):
  resultado=[]
  for i in range(len(cada)):
    s=0
    for j in range(len(cadb)):
      if cada[i]!=cadb[j]:
        s+=1
    if s==len(cadb):
      resultado.append(cada[i])
  for j in range(len(cadb)):
    resultado.append(cadb[j])
  return resultado

def diferencia(cada,cadb):
  resultado=[]
  for i in range(len(cada)):
    s=0
    for j in range(len(cadb)):
      if cada[i]!=cadb[j]:
        s+=1
    if s==len(cadb):
      resultado.append(cada[i])
  for j in range(len(cadb)):
    s=0
    for i in range(len(cada)):
      if cadb[j]!=cada[i]:
        s+=1
    if s==len(cada):
      resultado.append(cadb[j])
  return resultado

print(cada)
print(cadb)
print(union(cada,cadb))
print(diferencia(cada,cadb))
```
